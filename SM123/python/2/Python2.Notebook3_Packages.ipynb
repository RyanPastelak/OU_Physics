## 3.1   The numpy package

In Python 1 you were introduced to modules, and the fact that these contain useful functions, commands and constants. For example, you imported the module `math` and used its function `sin`. Section 2.5 of Python 1 Notebook 2 lists a number of functions available in the `math` module.

The package `numpy` contains a large number of extremely useful functions for scientific computing. The mathematical functions available in the `math` module are also available in the `numpy` package. Because some functions can be available in more than one module or package, it is sometimes important to specify which module or package a function comes from. Let’s look at the example below.

```python
# This program calculates the sine of an angle using numpy  functions

import numpy as np   # import the module called numpy and give it the name np

theta = float(input("Angle in degrees?"))  # ask user to input an angle in degrees
theta_rad = np.pi * theta/180              # convert the angle from degrees to radians by  
                                           # multiplying by pi and dividing by 180
result = np.sin(theta_rad)                 # take the sine of the angle in radians; store it in result

print ("sine of", theta,"is", result)
```

There are several things to note. In the first line, the package `numpy` is imported and given a short name `np`. This name is then used when both the constant $\pi$ and the sine function are called. That is, `np.pi`  and `np.sin` are used (and not just `pi` and `sin`). A different name could have been chosen and used, for example,

<code>import numpy as potatoes</code>

but then the sine function would have to be called `potatoes.sin` (clearly not a very wise choice as the name is quite long to write and not informative at all). Note also that the angle is converted from degrees to radians. This is because the function `sin`  expects the angle to be input in radians.

Note also that this `import` statement is different to what you saw in Python 1 Section 2.5. There the instruction was <code>from math import *</code> which imported all the functions from the `math` module with their existing names. Instead of that we could have written import `math` as `mt` and then referred to the various functions in the program as `mt.pi` and `mt.sin`, for instance.

## 3.2  The linspace function

Another useful function in `numpy` is `linspace`. This function returns evenly spaced numbers over a specified interval. For example, you may want all the integer numbers in the interval 1 to 10. Or you may need to use a list of numbers that covers the range 1.0 to 2.0 in steps of 0.05. The syntax to use is <code>linspace(start,stop,num,endpoint)</code>, where `start` and `stop` are the start and end values of the sequence of numbers, `num` is the number of steps in the sequence, and `endpoint` determines whether or not the end value of the sequence is included or not. Below is a program that uses this function to create two sets of numbers, and print them out.

```python
# This program generates two example arrays using the  
# function linspace from numpy

import numpy as np               # import the module called numpy and give it the name np

x=np.linspace(-10,10,20,endpoint=True)  # create array with 20 values between -10 and 10 and  
                                        # include the maximum value (10) 
y=np.linspace(-10,10,20,endpoint=False) # create array with 20 values between -10 and 10 but  
                                        # don't include the maximum value (10) 

print ("x=", x)
print ("y=", y)
```

The numbers are stored in arrays. Arrays are like lists, but they have different properties and you can perform different operations on them. (You will explore arrays further in Python 3.) In lines 6 and 7, the arrays `x` and `y` are each defined to contain 20 values ranging between −10 and 10. If you run the program, you will notice that all of the values in the array <code>x</code> are non-integer numbers with many decimal places, apart from the first and last values of −10.0 and +10.0. Conversely, those in array `y` are all integer values, but they only span the range from −10.0 to +9.0.

The difference is because of the final argument of the function called <code>endpoint</code>. As you can see, this can have one of two values: `True` or `False`. When `endpoint=True`, the maximum value of the range is included, so the largest value in the array `x` is indeed +10.0. However, there are 21 integers between −10 and +10 (including zero), not 20 integers. So dividing the full range from −10 to +10 into 20 steps results in the (mostly) non-integer values shown, Conversely, on line 7 where `endpoint=False`, the upper value of the range is not included, so the 20 values in the array y are all integers, ranging from −10.0 to +9.0.

Try changing line 6 to read <code>x=np.linspace(-10,10,21,endpoint=True)</code> and re-run the program. This will give what we want: 21 integer values between −10.0 and +10.0 inclusive.
