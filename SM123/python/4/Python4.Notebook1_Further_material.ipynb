#  Another way of  reading from a file

The folder <code>Python 4</code> in the OCL  contains a file named <code>Test_data.txt</code>. We are going to read this file:

```python
# Example of how to read the content of a file named Test_data.txt.

with open("Test_data.txt", "r") as input_file:       # open file from which data will be read
    input_data = input_file.read()                  # read and store data

print ('Initial data')                                 # print data
print (input_data) 
```

Line 3 does several things: the instruction `open` opens the file with the name indicated in quotation marks and gives it the name `input_file`. It opens the file in read mode (i.e. to read from it, but *not* write to it) as indicated by the `"r"`. Finally, the use of `with` ensures that the file is closed after the data is read. This is a bit of a technical point that you don't need to concern yourself with at this stage, but it is good practice to close files once you have finished reading from them and this syntax ensures that this is the case.

Line 4 reads all the data in the file and stores it in `input_data` as a string. Note something very important: line 4 is indented. The reading must be done while the file is still open, so it must be part of the  `with` block.  Line 5 prints  the data read from the file that we stored in `input_data`.  You can open the <code>Test_data.txt</code> file by clicking on it to check that what is printed above is exactly what the file contains.

If we want to plot or operate with the data we've read in, a string is not useful: we need to turn the data into lists or arrays.  To create one or more lists we need to split the string. This is done using the instruction `split`. In this case, we would like two lists: one with the numbers that appear on the left column printed above and another that contains the numbers on the right.  There are various ways to do this. 

The instruction `split` takes an argument between the brackets. I first split the string into the  elements that correspond to each of the lines  using `.split('\n')` Note, in line 11 of the example below, that you can just append `.split(argument)` to the definition of the string. Once I've done this, I want to split each elemnt in the list `number_list` into two numbers and put each in a different list. 

The program below  uses an iteration instruction (`for i`) to go over all the elements in `number_list`. It then checks that the string is not empty using if i (in other words, it only executes the following block of code if the string i contains something). Then, in a single line of code, the program splits each of these strings using `i.split('\t')` and puts the first part of the string into the variable `x` and the second part into the variable `F`. Finally it appends the strings `x` and `F` to the lists called `xlist` and `Flist`. In this case I use `\t`  because I want the string to be split where there were blank spaces in the initial string (in fact, I could use `.split( )` and this would also work. 

```python
# Example of how to read the content of a file and store it as lists.

with open("Test_data.txt", "r") as input_file:       # open file from which data will be read
    input_data = input_file.read()                  # read and store data

# Create the lists in which to store the values of force and extension 
xlist=[]
Flist=[]

number_list = input_data.split('\n')     # Split  input_data into several elements each containg a line
print ('Splitting by lines',number_list)

for i in number_list:     # iterate over all elements in number_list
    if i:
      x,F = i.split('\t')    # split each element in two called  x and F
      xlist.append(x)     # add the elements to the list extension
      Flist.append(F)     # add the elements to the list force
        
print (xlist)                                 # print data
print (Flist)                                 # print data
```

You can see that when we print `xlist` and `Flist` at the end of the program, we get two lists.
