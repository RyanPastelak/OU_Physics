# A possible answer to the Summary Activity in Python 1

```python
# This program calculates, for a  mass provided by the user, the kinetic energy  in 
# kilograms (metres per second)**2 for a number of speeds.
	
speeds = [1.0,2.0, 3.0, 4.0]

while True:
    answer = input("Mass of the object in kg?")
    if answer == "stop":
        break
    else:
        mass = float(answer)
        for speed in speeds:
          kinetic_e = mass*(speed**2)/2
          print ("Kinetic energy:", kinetic_e, "kilograms (metres per second)**2")
```
