# A possible answer to Activity 1.4 in Python 1

```python
totmass = 10  # total mass in kilograms 
latitude = 30  # latitude in degrees (N or S) 

if latitude < 60 and latitude > 30: # if the value of latitude is smaller than 60 AND 
                                    # bigger than 30 execute the line below
    acc = 9.81                      # acceleration in metres/(sec squared) 
elif latitude <= 30:                # if the value of latitude is smaller than 
                                    # or equal to 30 execute the line below
    acc = 9.78                      # acceleration in metres/(sec squared) 
else:                               # for all other values of the latitude excecute the line below
    acc = 9.83                      # acceleration in metres/(sec squared) 

weight = totmass * acc 

print ("Total weight:", weight, "newtons") 
```
